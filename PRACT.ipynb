```python
# image rotate syntax
# cv.getRotationMatrix2D(center,angle,scale)
# scale 
# 1.0 means same size
# 0.5 means half size
# 2.0 means double size
# cv.warpAffine(image,rotation_matrix,dimension)
# dimension means
# (width,height)
#image means
# image to be transformed
# rotation_matrix means
# matrix returned by getRotationMatrix2D
def rotate(img,angle,point=None):
    height,width=img.shape[:2]
    if point==None:
        point=img.shape[1]//2,img.shape[0]//2
    # point is center
    rotmat=cv.getRotationMatrix2D(point,angle,1.0)    
    dim=(width,height)
    return cv.warpAffine(img,rotmat,dim)
```

```python
cv.rectangle(blank,(0,0),(230,230),(0,0,0),thickness=-1)
plt.imshow(blank)
plt.axis('off')
plt.show()
```

```python
cv.rectangle(blank,(270,270),(500,500),(0,0,0),thickness=-1)
plt.imshow(blank)
plt.axis('off')
plt.show()
```

```python
ri=rotate(blank,45)
plt.imshow(ri)
plt.axis('off')
plt.show()
```

```python
cv.circle(ri,(250,250),60,(0,250,0),thickness=-1)
plt.imshow(ri)
plt.axis('off')
plt.show()
```

```python
cv.circle(ri,(250,250),59,(0,0,0),thickness=-1)
plt.imshow(ri)
plt.axis('off')
plt.show()
```

```python
nb=np.ones((600,600,3),dtype='uint8')
nb[:,:]=(250,250,250)
```

```python
plt.imshow(nb)
plt.axis('off')
plt.show()
```

```python
cv.circle(nb,(nb.shape[0]//2,nb.shape[1]//2),250,(0,250,0),thickness=-1)
plt.imshow(nb)
plt.axis('off')
plt.show()
```

```python
def draw_sector(img,start_angle,end_angle,thickness,radius,center,angle):
    height,width=img.shape[:2]
    axis=(radius,radius)
    return cv.ellipse(img,center,axis,angle,start_angle,end_angle,(250,250,250),thickness)
img=draw_sector(nb,135,215,-1,280,(nb.shape[0]//2+30,nb.shape[1]//2),0)
# cv.imshow("dhc",img)
img=draw_sector(img,135,215,-1,280,(nb.shape[0]//2-30,nb.shape[1]//2),180)
plt.imshow(img)
plt.axis('off')
plt.show()
    
```
